# This code uses Voting Classifier in conjunction with Grid Search.

We are using 3 classifiers - 

1. Logistic Regression
2. Random Forest Classifier
3. Support Vector Classifier

On top of these, we iterate over the following parameters for the models - 
1. Logistic Regression : C
2. Random Forest : Max Depth
3. SVC : Gamma
    
This gives us the advantage of having multiple classifiers as well as iterations over multiple parameters. 
Hence we account for multiple learners with inputs from weak and strong learners both.


```python
#Import relevant libraries
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import metrics 
    
#Load Iris Dataset
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                    test_size=0.3, random_state=0)

#Create list of tuples of all the classifiers and their aliases
estimators = [('lr', LogisticRegression()), ('rf', RandomForestClassifier()), ('svc', SVC(probability = True))]

#Create instance of Voting Classifier class and pass estimator list.
#We use Voting method as Soft rather than Hard.
########### Hard - Take individual model output (1 or 0) and use that for voting
########### Soft - Take argmax of individual probabilities for voting
clf = VotingClassifier(estimators, voting = 'soft')


#Create a dictionary of all the parameters we need to Grid Search on. 
#Convention is to use model alias, then __ (dunder) and then exact parameter name (Case Sensitive)
#Pass the list of all the values to be tested as the value for the above key
params = {'lr__C' : [1,5,10],
          'rf__max_depth' : [2,3,4],
          'svc__gamma' : ['scale', 'auto']}

#Initialize GridSearchCV method
grid = GridSearchCV(clf, params, cv = 5)

#Fit training data on the Grid Search performed on Voting Classifier
grid.fit(X_train, y_train)

#Use fitted Grid to perform predictions on test data
y_pred = grid.predict(X_test)

#Get Model Score
if y_pred is not None:
    print(metrics.accuracy_score(y_test, y_pred))

#We can check which combination gives maximum accuracy
grid.best_params_
```
